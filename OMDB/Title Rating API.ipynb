```python
#import dependencies to read the files 
import os 
import csv 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import json
import requests
from config import api_key
```

```python
from pprint import pprint
```

```python
# File to Load
file_to_load = "../netflix_titles_CLEANSEDcsv.csv"

# Read Purchasing File and store into Pandas data frame
Netflix_Titles = pd.read_csv(file_to_load)
```

```python
Netflix_Titles= Netflix_Titles.dropna()
```

```python
url = "http://www.omdbapi.com/?t="
api_key = "&apikey=" + api_key
```

```python
imdb= []
for titles in Netflix_Titles["title"]:
    imdb.append(titles)
```

```python
response = requests.get(url + "Da Kath & Kim Code" + api_key)

data = response.json()
pprint(data)
```

```python
url2 = "http://www.omdbapi.com/?apikey=" + api_key + "&t="
responses=[]
movie_title= []
imdb_rating= []
movie_genre= []
for movie in imdb:
    try: 
        movie_data = requests.get(url + movie + api_key).json()
        responses.append(movie_data)
        movie_title.append(movie_data["Title"])
        imdb_rating.append(movie_data['imdbRating'])
        movie_genre.append(movie_data["Genre"])
        print(f"The Netflix Title {movie_data['Title']}.")
        print(f"Received an IMDB rating of {movie_data['imdbRating']}.")
        print(f"And falls under the {movie_data['Genre']} genre.")
    except(KeyError, IndexError):
        print("Missing Results.")
        
    print("------------")
    
```

```python
#put api results into a dataframe
imdb_rating_df= pd.DataFrame({
    "Title": movie_title,
    "IMDB Rating": imdb_rating,
    "Genre": movie_genre
})

#drop any missing values
imdb_rating_df= imdb_rating_df.dropna(how ="any")
imdb_rating_df
```

```python
imdb_rating_df.to_csv("Movie IMDB Rating.csv", index= False)
```
